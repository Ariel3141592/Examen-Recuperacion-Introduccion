#**Top 5 Ciudades/Sectores Más Caros**

```python
Alquiler_por_Ciudad = df.groupby("Provincia")["Precio"].mean()
print("El Ranking de las 5 Propiedades mas Caras por Ubicacion es:")
Alquiler_por_Ciudad_descendente = Alquiler_por_Ciudad.sort_values(ascending=False)
Alquiler_por_Ciudad_descendente.head()
```
