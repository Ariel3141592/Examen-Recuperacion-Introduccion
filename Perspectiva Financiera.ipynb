```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
df = pd.read_csv("real_estate_ecuador_200k.csv")
df
```

#**Costo promedio de Alquiler**

```python
Alquiler_general = df["Precio"].mean()
print("El costo de alquiler general es:")
print(Alquiler_general)
```

```python
Alquiler_por_Ciudad = df.groupby("Provincia")["Precio"].mean()
print("El costo de alquiler por Provincia es:")
Alquiler_por_Ciudad
```

# **Costo Promedio por Metro Cuadrado**

```python
df['Costo por Metro Cuadrado'] = df['Precio'] / df['Area']
Costo_Metro_Cuadrado_general = df['Costo por Metro Cuadrado'].mean()
print("El costo promedio por metro cuadrado es:")
print(Costo_Metro_Cuadrado_general)
```

# **Rango de Precios (Min/Max/Media) por Provincia**

```python
Minimo_por_provincia = df.groupby('Provincia')['Precio'].min()
print(Minimo_por_provincia)
```

```python
Maximo_por_provincia = df.groupby('Provincia')['Precio'].max()
print(Maximo_por_provincia)
```

```python
Promedio_por_provincia = df.groupby('Provincia')['Precio'].mean()
print(Promedio_por_provincia)
#
```
